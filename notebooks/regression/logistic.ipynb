```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import seaborn as sns
import sys

sys.path.append("../../")

from mlalgorithms.linear_model import LogisticRegression
from mlalgorithms.utils import train_test_split, normalize
from mlalgorithms.metrics import mean_squared_error

from sklearn.metrics import accuracy_score

%matplotlib inline
```

```python
X, y = load_breast_cancer(return_X_y=True)
# X = normalize(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

print("Train: ", X_train.shape)
print("Test: ", X_test.shape)
```

```python
clf = LogisticRegression(eta=0.001, n_iters=200, C=0.03, penalty='L1')
t, v = clf.fit(X_train, y_train)
plt.plot(range(len(t)), t, 'g-', label='Training Loss')
plt.plot(range(len(v)), v, 'y--', label='Validation Loss')
plt.title("Learning Curve")
plt.legend(loc='best')
plt.show()
```

```python
y_pred = clf.predict(X_test)

cutoff = 0.5

y_pred[y_pred > cutoff] = 1
y_pred[y_pred <= cutoff] = 0

print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression


sclf = LogisticRegression()
sclf.fit(X_train, y_train)

y_pred = sclf.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
```
