```python
import sys
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

sys.path.append("../../")

from mlalgorithms.regression import OLSRegression, GradientRegression
```

```python
X, y = make_regression(n_samples=500, n_features=1, bias=0.5, noise=20, random_state=42)
plt.plot(X, y, 'ro')
plt.xlabel("$X$")
plt.ylabel("$y$")
plt.title("Regression Data Sample")
plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
model = OLSRegression()
model.fit(X, y)
```

```python
plt.plot(X_test, y_test, 'ro')
plt.plot(X_test, model.predict(X_test), 'b--')
plt.xlabel("$X$")
plt.ylabel("$y$")
plt.title("OLS Regression Model")
plt.show()
```

```python
model1 = GradientRegression(n_iters=100, learning_rate=0.03)
model1.fit(X_train, y_train)
```

```python
X, y = make_regression(n_samples=100, n_features=13, noise=20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4) 
```

```python
mv_model = OLSRegression(C=0.01)
mv_model.fit(X_train, y_train)
```

```python
mv_model1 = GradientRegression(learning_rate=0.01, regularize=False)
mv_model1.fit(X_train, y_train)
```

```python
print("OLS test MSE: ", mv_model.mean_squared_error(X_test, y_test))
print("Gradient test MSE: ", mv_model1.mean_squared_error(X_test, y_test))
```

```python
print("OLS test R2 score: ", mv_model.r2_score(X_test, y_test))
print("Gradient test R2 score: ", mv_model1.r2_score(X_test, y_test))
```
